### Identificando a correlação entre s variáveis

### **CORRELAÇÃO NÃO É CAUSALIDADE!!**

```python
def plot_corr(df, size=10):
    corr = df.corr()
    fig, ax = plt.subplots(figsize = (size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
plot_corr(dataset)
```

### Visualizando a correlação em tabela númerica 

Coeficiente de correlação:

     +1 = forte correlação positiva
      0 = não há correlação
     -1 = forte correlação negativa

### Definindo as classes

```python
diabetes_map = {True: 1, False : 0}
```

### Aplicando o mapeamento ao dataset

```python
dataset["diabetes"] = dataset["diabetes"].map(diabetes_map)
```

```python
dataset.head(10)
```

### Verificando como os dados estão distribuidos

```python
num_true = len(dataset.loc[dataset["diabetes"] == True])
num_false = len(dataset.loc[dataset["diabetes"] == False])
print(f"Número de casos verdadeiros: {num_true} ({num_true / (num_true + num_false) * 100})")
print(f"Número de casos falsos     : {num_false} ({num_false / (num_true + num_false) * 100})")
```

### Separação dos dados em TREINO e TESTE

```python
from sklearn.model_selection import train_test_split
```

```python
# Seleção de variáveis preditoras (Feature Selection)
atributos = ['num_preg', 'glucose_conc', 'diastolic_bp', 'thickness', 'insulin', 'bmi', 'diab_pred', 'age']
```

```python
# Variável a ser prevista
atrib_prev = ['diabetes']
```

```python
# Criando objetos
X = dataset[atributos].values
Y = dataset[atrib_prev].values
```

```python
X, Y
```

```python
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size= 0.20, random_state=42)
```

```python
# Imprimindo os resultados
print("{0:0.2f}% nos dados de treino".format((len(X_treino)/len(dataset.index)) * 100))
print("{0:0.2f}% nos dados de teste".format((len(X_teste)/len(dataset.index)) * 100))
```

```python
X_treino, Y_treino
```

```python
X_treino.shape, Y_treino.shape
```

```python
X_teste.shape, Y_teste.shape
```

### Verificando o Split (Divisão)

```python
print("Original True : {0} ({1:0.2f}%)".format(len(dataset.loc[dataset['diabetes'] == 1]), 
                                               (len(dataset.loc[dataset['diabetes'] ==1])/len(dataset.index) * 100)))

print("Original False : {0} ({1:0.2f}%)".format(len(dataset.loc[dataset['diabetes'] == 0]), 
                                               (len(dataset.loc[dataset['diabetes'] == 0])/len(dataset.index) * 100)))
print("")
print("Training True : {0} ({1:0.2f}%)".format(len(Y_treino[Y_treino[:] == 1]), 
                                               (len(Y_treino[Y_treino[:] == 1])/len(Y_treino) * 100)))

print("Training False : {0} ({1:0.2f}%)".format(len(Y_treino[Y_treino[:] == 0]), 
                                               (len(Y_treino[Y_treino[:] == 0])/len(Y_treino) * 100)))
print("")
print("Test True : {0} ({1:0.2f}%)".format(len(Y_teste[Y_teste[:] == 1]), 
                                               (len(Y_teste[Y_teste[:] == 1])/len(Y_teste) * 100)))

print("Test False : {0} ({1:0.2f}%)".format(len(Y_teste[Y_teste[:] == 0]), 
                                               (len(Y_teste[Y_teste[:] == 0])/len(Y_teste) * 100)))
```

### Valores Missing Ocultos

```python
### Valores Missing Ocultos
print("# Linhas no dataframe {0}".format(len(dataset)))
print("# Linhas missing glucose_conc: {0}".format(len(dataset.loc[dataset['glucose_conc'] == 0])))
print("# Linhas missing diastolic_bp: {0}".format(len(dataset.loc[dataset['diastolic_bp'] == 0])))
print("# Linhas missing thickness: {0}".format(len(dataset.loc[dataset['thickness'] == 0])))
print("# Linhas missing insulin: {0}".format(len(dataset.loc[dataset['insulin'] == 0])))
print("# Linhas missing bmi: {0}".format(len(dataset.loc[dataset['bmi'] == 0])))
print("# Linhas missing age: {0}".format(len(dataset.loc[dataset['age'] == 0])))
```

### Fazendo o tratamento dos dados Missing - Impute

Sustituindo os valores iguais a zero, pela média dos dados

```python
from sklearn.impute import SimpleImputer
```

```python
# Criando o Objeto para preencher os valores zeros com a média
preenche_0 = SimpleImputer(missing_values = 0, strategy="mean")

# Substituindo os valores iguais a zero, pela média dos dados
X_treino = preenche_0.fit_transform(X_treino)
X_teste = preenche_0.fit_transform(X_teste)
```

### Construindo e treinando o modelo

Utilizando o classificador **Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
```

```python
modelo_v1 = GaussianNB()
modelo_v1.fit(X_treino, Y_treino.ravel())
```

Verificando a exatidão no modelo nos dados de treino

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
nb_predict_train = modelo_v1.predict(X_treino)
accuracy_score(Y_treino, nb_predict_train)
```

Verificando a exatidão nos dados de teste

```python
nb_predict_test = modelo_v1.predict(X_teste)
accuracy_score(Y_teste, nb_predict_test)
```

Métricas

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm = ConfusionMatrix(modelo_v1)
cm.fit(X_treino, Y_treino)
cm.score(X_teste, Y_teste)
```

```python
print(classification_report(Y_teste, nb_predict_test))
```

***

Otimizando o Modelo com **RandomForest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_v2 = RandomForestClassifier(random_state=42)
modelo_v2.fit(X_treino, Y_treino.ravel())
```

Verificando os dados de treino

```python
rf_predict_train = modelo_v2.predict(X_treino)
print(accuracy_score(Y_treino, rf_predict_train))
```

Verificando os dados de teste

```python
rf_predict_test = modelo_v2.predict(X_teste)
print(accuracy_score(Y_teste, rf_predict_test))
```

```python
cm = ConfusionMatrix(modelo_v2)
cm.fit(X_treino, Y_treino)
cm.score(X_teste, Y_teste)
```

```python
print(classification_report(Y_teste, nb_predict_test))
```

***

Otimizando o Modelo com **Regressão Logística**

```python
from sklearn.linear_model import LogisticRegression
```

```python
modelo_v3 = LogisticRegression(C = 0.7, random_state= 42, max_iter = 1000)
modelo_v3.fit(X_treino, Y_treino.ravel())
lr_predict_test = modelo_v3.predict(X_teste)
```

```python
cm = ConfusionMatrix(modelo_v3)
cm.fit(X_treino, Y_treino)
cm.score(X_teste, Y_teste)
```

```python
print(classification_report(Y_teste, lr_predict_test))
```
